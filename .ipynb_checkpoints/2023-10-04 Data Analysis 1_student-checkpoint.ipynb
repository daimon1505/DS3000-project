For convenience, let's write it as a function that takes a filter string as an argument, loads all files that match the filter and returns a pandas dataframe that contains all the loaded information.


```python
def load_csvs(path, fname_filter):
    files = glob.glob(f'{path}/{fname_filter}')
    return pd.concat([pd.read_csv(f) for f in files], ignore_index=True)
```

### Boolean operations
We can also do binary operations like: `&` (and), `|` (logical-or), as well as unary operators like `~` (logical not) on pandas series.

A few examples:
* How many times a product was marked_as_amazon **and** also appeared in the Amazon only searches?

```python
((products_all_df['marked_as_amazon']==True) & (products_all_df['in_amazon_brands']==True)).sum()
```

Because these are boolean columns, we can also skip `==True` and get the same effect:

```python
((products_all_df['marked_as_amazon']) & (products_all_df['in_amazon_brands'])).sum()
```

* Are there any products that aren't marked as amazon brand in the search results but still appear in the amazon only search?

```python
((~products_all_df['marked_as_amazon']) & (products_all_df['in_amazon_brands'])).sum()
```

Notice the use of brackets. They are not always necessary but when they are and you forget them you will get very wrong results and likely not even realize it. Compare the two lines below trying to answer how often amazon products are rated higher than 4.5:

```python
(products_all_df['marked_as_amazon'] & products_all_df['rating'] > 4.5).sum()
```

```python
(products_all_df['marked_as_amazon'] & (products_all_df['rating'] > 4.5)).sum()
```

What happened?!

### Accessing rows using boolean criteria

In these cases, we will be using `.loc` instead of `.iloc`. Remember:
* `.loc` is for finding rows and/or columns with particular labels
* `.iloc` is for finding rows and/or columns located at a particular integer location

See [this StackOverflow answer](https://stackoverflow.com/a/31593712/18264897) for more details.

## Recreating the analysis from The Markup
Ok but how do we actually prove that it's amazon and not higher ratings or more reviews?
Let's continue re-creating the method from The Markup.

### Preparing the data
Remember, they look at the first two results for each search and represent their differences as one row:

![features](https://mrkp-static-production.themarkup.org/graphics/amazon-methodology-product-comparison/1634143723511/assets/product-pairs-subtraction-desktop.png)

We can use the .isin() to only select the top two rows from our results:

```python
topdf = products_all_df[products_all_df['rank'].isin([0,1])]
```

```python
topdf
```

```python
import random

# create a list for holding our product pairs
pairs = []

# we go from the beginning to the end of the top_df data frame but only every second row
for index in range(0, len(topdf), 2):

    # coin flip - who gets to be product A, who gets to be product B?
    if random.random() < 0.5:
        # accessing a row by numerical index with .iloc
        row = topdf.iloc[index]
        comparison = topdf.iloc[index+1]
    else:
        row = topdf.iloc[index+1]
        comparison = topdf.iloc[index]

    # create the pair as a dictionary
    pair = {}

    # is the product A placed higher (as in on top instead of as the second?)
    pair['is_placed_higher'] = (row['rank'] == 0)

    # now, for each variable let's calculate the delta
    for k in ['is_ad', 'rating', 'review_count', 'is_amazon']:
        pair[f'{k}_delta'] = float(row[k]) - float(comparison[k])
    pairs.append(pair)
    
    
    products_all_df.groupby('rank').mean
```

```python
pairs_dataset = pd.DataFrame(pairs)
pairs_dataset
```

Great! We have a dataset where every row represents a pair of products. 

The Markup then used RandomForests to try to predict which of the two products comes on top.
They chose RandomForests because this model reveals its trained "features importances". Because of that RandomForests models are more *explainable* than some other, potentially more predictive approaches without this insight.

#### Train/test split
Regardless of the method, in Machine Learning it is a standard practice to divide the datasets into training and test (validation) sets. The models are trained on the train set, which is usually the bigger set, and then they are asked to predict the validation (smaller) set that they had not seen before.

This is important, because the more complex a Machine Learning model is, the more likely it is to "overfit", i.e. fit traning data perfectly but become incapable of reliably predicting previously unseen data. Spliting into train and validation does not fix this problem, but it allows us to measure the model performance on the unseen data (the validation set).

#### Stratification
We `stratify` the train/test split to ensure that the proportions of the outcomes we are predicting are the same in the train and test set.

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(pairs_dataset, 
                               test_size=0.2, 
                               stratify=pairs_dataset['is_placed_higher'])
```

```python
# these are the features we will use for prediction

features = ['is_ad_delta', 'rating_delta', 'review_count_delta', 'is_amazon_delta']

from sklearn.ensemble import RandomForestClassifier


clf = RandomForestClassifier(n_estimators=100, # how many trees?
                             max_depth=3, # the deeper the tree the more accuracy in training but more potential for overfitting
                             max_features = 3) # the more features more accuracy in training but more potential for overfitting

clf.fit(train[features], train['is_placed_higher'])
```

If you got an error because of the presence of missing values (NaNs), we will need to remove those from our data using `.dropna()` and then go back to the train/test split:

```python
pairs_dataset.dropna(inplace=True)
```

We've now trained the model, let's see how well it predicts previously unseen data:


```python
# 1. predict the test data
y_pred = clf.predict(test[features])

# 2. calculate accuracy, i.e. the fraction of time that the prediction was correct out of all predictions
(y_pred == test['is_placed_higher']).sum()/len(y_pred)
```

```python
for feature, importance in zip(features, clf.feature_importances_):
    print(feature, importance)
```

And there we have it! We see that the most useful feature in predicting which product was on top is knowing whether it came from amazon.
That feature is more predictive than the difference in reviews and ratings!

There are still some limitations to interpreting the feature importances. We will cover both these shortcomings and how to address them on Friday.
